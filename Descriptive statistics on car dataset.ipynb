```python
#Central Tendencies Calculation
```

```python
print('MEAN: \n', car_df.mean())
print('\r \n')

print('MEDIAN: \n', car_df.median())
print('\r \n')

print('MODE: \n', car_df.mode())
print('\r \n')

print('Q1-Quartile 1: \n', car_df.quantile(q = (0.25, 0.5, 0.75)))
```

```python
# check if the column Selling_Price is normally distributed
```

```python
mean = car_df['Selling_Price'].mean()
median = car_df['Selling_Price'].median()
mode = car_df['Selling_Price'].mode()

fig, ax = plt.subplots(figsize = (10, 6))

# setting size of the canvas
sns.distplot(car_df['Selling_Price']);

plt.title('Distribution plot of Selling Price of Cars')

plt.axvline(mean, color = 'green', label='Mean')
plt.axvline(median, color = 'blue', label='Median')
plt.axvline(mode[0], color = 'red', label='Mode1')
plt.axvline(mode[1], color = 'red', label='Mode2')

plt.legend()
```

```python
# Checking Correlation
```

```python
corr = car_df.corr()
print(corr)

# sns.heatmap(corr, annot = True)

fig, ax = plt.subplots()
fig.set_size_inches(10, 10)

plt.title('Heat Map of Correlation \n')

sns.heatmap(corr, annot = True,  vmin = 1, vmax = 1, linewidths = .5, center = 0)
```
